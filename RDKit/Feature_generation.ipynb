### Converting compound names to smiles

Using CACTUS (which uses the NCI database) we can find matching smiles for our compound names

```python
from urllib.request import urlopen
from urllib.parse import quote

def to_smiles(name):
    try:
        url = 'http://cactus.nci.nih.gov/chemical/structure/' + quote(name) + '/smiles'
        smiles = urlopen(url).read().decode('utf8')
        return smiles
    except:
        return 'Did not work'
```

```python
df['Smiles'] = [to_smiles(item) for item in df['Organic']]
df['Add charge'] = False
df['Smiles'].value_counts(normalize = True)
```

```python
df
```

This function will serve to assess the percent of names successfully converted to smiles

```python
def percent_converted(smiles):
    not_converted = 0
    total = 0
    for item in smiles:
        if item == 'Did not work':
            not_converted += 1
            total += 1
        else:
            total += 1
    converted = (1-(not_converted/total))*100
    return converted    
```

As we see, the initial conversion is not ideal. Upon observation, some trends can be noted that lead to compound names not successfully converting. These include cations (CACTUS has many more neutral compounds than charged species), as well as simple nomenclature changes.

```python
print(percent_converted(df['Smiles']))
```

```python
for x in range(0, 760):
    compound = str(df.loc[x]['Organic'])
    smiles = str(df.loc[x]['Smiles'])
    if smiles == 'Did not work':
        print('Unchanged: ', compound)
        compound = compound.replace('-O-', '-2-')
        compound = compound.replace('-o-', '-2-')
        compound = compound.replace(' deprotonated', '')
        if 'ammonium' in compound:
            compound = compound.replace('ammonium', 'amine')
            add_charge = True   
        elif 'protonated' in compound:
            compound = compound.replace('protonated', '')
            add_charge = True
        if 'aminee' in compound:
            compound = compound.replace('aminee', 'aminoe')    
        print('Changed: ', compound)
        deprotonated_smile = to_smiles(compound)
        print(deprotonated_smile, 'Charge to add: ', add_charge)
        df.loc[x, 'Smiles'] = deprotonated_smile
        df.loc[x, 'Add charge'] = add_charge
        #added_charge.append(charge_to_add)
```

By applying our changes to the compounds that did not convert, we have increased the conversion from ~75% to ~92%

```python
print(percent_converted(df['Smiles']))
```

We need to re-protonate compounds that we manually deprotonated for the sake of converting to smiles. The below function accomplishes this.

```python
def protonate_amine(mol, protonate):
    if protonate == True:
        Nitrogen_index = mol.GetSubstructMatch(Chem.MolFromSmarts('N'))
        mol.GetAtomWithIdx(*Nitrogen_index).SetFormalCharge(1)
    return mol
```

## Generating 2D features

```python
molecules = [Chem.MolFromSmiles(item) for item in df['Smiles']]
add_charge = df['Add charge'].tolist()

#Create a dictionary containing molecule: charge_add pairs
charges = {molecules[i]: add_charge[i] for i in range(len(add_charge))} 

protonated_molecules = [protonate_amine(molecule, charges[molecule]) for molecule in molecules]

#Generate 2D features for each compound
df['Mol. Wt.'] = [Descriptors.MolWt(i) for i in protonated_molecules]
df['H-bond Donors'] = [Descriptors.NumHDonors(i) for i in protonated_molecules]
df['PSA'] = [Descriptors.TPSA(i) for i in protonated_molecules]
df['ASA'] = [Descriptors.LabuteASA(i) for i in protonated_molecules]
df['MolLogP'] = [Descriptors.MolLogP(i) for i in protonated_molecules]
df['Rotatable Bonds'] = [Descriptors.NumRotatableBonds(i) for i in protonated_molecules]
```

## Visualizing molecules

```python
def get_unique_mols(mols):
    unique_molecules = []
    unique_molecules_mw = []
    for item in mols:
        item_mw = Descriptors.MolWt(item)
        if item_mw not in unique_molecules_mw:
            unique_molecules.append(item)
            unique_molecules_mw.append(item_mw)
    return unique_molecules

uniques = get_unique_mols(protonated_molecules)
unique_smiles = [Chem.MolToSmiles(mol) for mol in uniques]
Draw.MolsToGridImage(uniques, legends=unique_smiles,  molsPerRow=4,subImgSize=(400,200), maxMols = len(uniques))
```

## Checking a new compound in the dataset

The functions below allow for searching the database to see if a structure is present. Alternatively, if the compound is not persent, the most structurally similiar structure is returned.

```python
def find_nearest_match(smiles, greatest_similarity = 0, nearest_match = 'None'):
    search_mol = Chem.MolFromSmiles(smiles)
    search_fingerprint = Chem.RDKFingerprint(search_mol)
    for item in df['Smiles']:
        item_as_mol = Chem.MolFromSmiles(item)
        item_as_fingerprint = Chem.RDKFingerprint(item_as_mol)
        similarity = DataStructs.FingerprintSimilarity(search_fingerprint,item_as_fingerprint)
        if similarity > greatest_similarity:
            greatest_similarity = round(similarity, 3)
            nearest_match = item
    if similarity == 1:
        return_phrase = "This molecule is present in the dataset"
    else:
        return_phrase = "The nearest match is " + str(nearest_match) + " with a Tanimoto similarity of " + str(greatest_similarity)
    print(return_phrase)
    
def nearest_match(smiles, greatest_similarity = 0, nearest_match = 'None'):
    search_mol = Chem.MolFromSmiles(smiles)
    search_fingerprint = Chem.RDKFingerprint(search_mol)
    for item in df['Smiles']:
        item_as_mol = Chem.MolFromSmiles(item)
        item_as_fingerprint = Chem.RDKFingerprint(item_as_mol)
        similarity = DataStructs.FingerprintSimilarity(search_fingerprint,item_as_fingerprint)
        if similarity > greatest_similarity:
            greatest_similarity = round(similarity, 3)
            nearest_match = item
    return (str(nearest_match), str(greatest_similarity))
    
def match_printout(smiles):
    nearest_smiles = (nearest_match(smiles)[0])
    near_match = (nearest_match(smiles)[1])
    molecule1 = Chem.MolFromSmiles(smiles)
    molecule2 = Chem.MolFromSmiles(nearest_smiles)
    if near_match == '1.0':
        legend_phrase = str('Exact match. ' + smiles + ' is currently present in the dataset.')
        print(legend_phrase)
        return Draw.MolToImage(molecule1)
    else:
        legend_phrase = str('The nearest match to ' + smiles + ' (left) in the dataset is ' + nearest_smiles + ' (right).\nSimilarity index: ' + near_match)
        print(legend_phrase)
        return Draw.MolsToGridImage((molecule1, molecule2), molsPerRow=2,subImgSize=(200,200))
    
```

```python
pyridine_smiles = to_smiles('pyridine')
find_nearest_match(pyridine_smiles)
```

```python
match_printout(pyridine_smiles)
```

```python
diisopropylamine_smiles = to_smiles('diisopropylamine')
match_printout(diisopropylamine_smiles)
```

#  3D Visualization (demonstrated on rdkit.blogspot)

```python
import py3Dmol
def drawmodel(mol,p=None,confId=-1):
        molblock = Chem.MolToMolBlock(mol,confId=confId)
        if p is None:
            p = py3Dmol.view(width=400,height=400)
        p.removeAllModels()
        p.addModel(molblock,'sdf')
        p.setStyle({'stick':{}})
        p.setBackgroundColor('white')
        p.zoomTo()
        return p.show()
```
